# data cleaning

```python
boxplot_data = [df[df['area_type'] == area_type]['price'] for area_type in df['area_type'].unique()]

# Step 3: Create the boxplot with logarithmic scaling
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
plt.boxplot(boxplot_data, labels=df['area_type'].unique())
plt.xlabel('Area Type')
plt.ylabel('Price')
plt.yscale('log')  # Apply logarithmic scaling to the y-axis
plt.title('Boxplot of Prices in Different Area Types (Log Scale)')
plt.show()
```
