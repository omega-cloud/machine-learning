# data visualization(step4)

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
 df.drop('society', axis=1, inplace=True)
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df['balcony'].fillna(df['balcony'].mean(), inplace=True) # using mean to replace nan value of balcony 
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.dropna(inplace=True) # cleaning all the remaining null values 
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
# Boxplot for 'area_type' vs. 'price'
plt.figure(figsize=(10, 6))
sns.boxplot(x='area_type', y='price', data=df)
plt.title('Area Type vs. Price')
plt.xlabel('Area Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot for 'total_sqft' vs. 'price'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_sqft', y='price', data=df)
plt.title('Total Square Feet vs. Price')
plt.xlabel('Total Square Feet')
plt.ylabel('Price')
plt.show()
```

```python

# Boxplot for 'bath' vs. 'price'
plt.figure(figsize=(10, 6))
sns.boxplot(x='bath', y='price', data=df)
plt.title('Number of Bathrooms vs. Price')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot for 'balcony' vs. 'price'
plt.figure(figsize=(10, 6))
sns.boxplot(x='balcony', y='price', data=df)
plt.title('Number of Balconies vs. Price')
plt.xlabel('Number of Balconies')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Drop rows with non-numeric values in 'total_sqft' column
df = df[pd.to_numeric(df['total_sqft'], errors='coerce').notnull()]
```

```python
# Visualizing the correlation between numerical features using a heatmap
# Consider selecting the relevant numerical columns for correlation analysis

numerical_cols = ['total_sqft', 'bath', 'balcony', 'price']  # Update with relevant columns
```

```python
corr_matrix = df[numerical_cols].corr()

# Generate a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
